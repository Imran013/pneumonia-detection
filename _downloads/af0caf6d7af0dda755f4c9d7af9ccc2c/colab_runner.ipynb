```python
import re

with open('config.py', 'r') as f:
    config = f.read()

# GPU settings
config = re.sub(r"device: str = 'cpu'", "device: str = 'cuda'", config)
config = re.sub(r"num_workers: int = 0", "num_workers: int = 2", config)
config = re.sub(r"pin_memory: bool = False", "pin_memory: bool = True", config)

with open('config.py', 'w') as f:
    f.write(config)

print("âœ… Config updated for GPU!")
```
